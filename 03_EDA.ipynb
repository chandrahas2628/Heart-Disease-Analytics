```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

# Load data
df = pd.read_csv("../data/heart_cleaned.csv")

# Dataset overview
print("Dataset Shape:", df.shape)
print("\nAverage Age:", round(df["age"].mean(), 2))
print("\nAverage BMI:", round(df["BMI"].mean(), 2))
print("\nAverage Cholesterol:", round(df["totChol"].mean(), 2))

print("\nCHD Distribution:")
print(df["TenYearCHD"].value_counts())

# -----------------------------
# CHD Distribution
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="TenYearCHD", data=df)
plt.title("10-Year Heart Disease Risk")
plt.savefig("../images/chd_distribution.png")
plt.show()

# -----------------------------
# Age Distribution
# -----------------------------
plt.figure(figsize=(8,5))
sns.histplot(df["age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.savefig("../images/age_distribution.png")
plt.show()

# -----------------------------
# Gender vs CHD
# -----------------------------
plt.figure(figsize=(8,5))
sns.countplot(x="male", hue="TenYearCHD", data=df)
plt.title("Gender vs CHD")
plt.savefig("../images/gender_vs_chd.png")
plt.show()

# -----------------------------
# Smoking vs CHD
# -----------------------------
plt.figure(figsize=(8,5))
sns.countplot(x="currentSmoker", hue="TenYearCHD", data=df)
plt.title("Smoking vs CHD")
plt.savefig("../images/smoking_vs_chd.png")
plt.show()

# -----------------------------
# Diabetes vs CHD
# -----------------------------
plt.figure(figsize=(8,5))
sns.countplot(x="diabetes", hue="TenYearCHD", data=df)
plt.title("Diabetes vs CHD")
plt.savefig("../images/diabetes_vs_chd.png")
plt.show()

# -----------------------------
# Systolic BP vs CHD
# -----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x="TenYearCHD", y="sysBP", data=df)
plt.title("Systolic BP vs CHD")
plt.savefig("../images/sysbp_vs_chd.png")
plt.show()

# -----------------------------
# Cholesterol vs CHD
# -----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x="TenYearCHD", y="totChol", data=df)
plt.title("Cholesterol vs CHD")
plt.savefig("../images/cholesterol_vs_chd.png")
plt.show()

# -----------------------------
# BMI vs CHD
# -----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x="TenYearCHD", y="BMI", data=df)
plt.title("BMI vs CHD")
plt.savefig("../images/bmi_vs_chd.png")
plt.show()
```
